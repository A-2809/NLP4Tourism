```python
# #dropping rows with nan values
df = train_df.dropna()
```

```python
#spliting the database
from sklearn.model_selection import train_test_split
x=df.loc[:,['Total tourism employment (direct) as % of total employment', 'GDP per capita','Pollution','Total international expenditure','Goverment Effectiveness', 'Total Domestic Trip', 'Rule of Law', 'Percent of total protected land','Exchange rate','Trade%GDP', 'Women Buisness and Law index',	'Hospital bed'	, 'Regulatory quality']]
y=df.loc[:,'Total international arrivals']
```

```python
from sklearn.preprocessing import MinMaxScaler #fixed import
scaler = MinMaxScaler()
X = scaler.fit_transform(x)
```

```python
#Splitting the dataset into 80-20 
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20, random_state=2)
```

```python
# Regression
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.tree import DecisionTreeRegressor 
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error


clf1 = LinearRegression() 
clf2 = SVR(kernel='poly')
clf3 = RandomForestRegressor(max_depth=10, n_estimators=100, random_state=2)
clf4 = KNeighborsClassifier(n_neighbors=2)
clf5 = MultinomialNB()
clf6 = DecisionTreeRegressor(random_state=10,max_features='sqrt')

clf1.fit(x_train, y_train)
predicted = clf1.predict(x_test)

print('MAE: ', metrics.mean_absolute_error(y_test, predicted))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predicted)))
print("MAPE:", metrics.mean_absolute_percentage_error(y_test, predicted))
print("R2 Score: ",metrics.r2_score(y_test, predicted)) 
print(1-(1-r2_score(y_test, predicted))*((len(x_test)-1)/(len(x_test)-len(x_test[0])-1)))
```

```python
import lightgbm as lgb

train_data= lgb.Dataset(data= x_train, label=y_train, feature_name = list(x) )
valid_data=lgb.Dataset(data= x_test, label=y_train, feature_name= list(y) )
parameters= {'objective': 'regression'}

model_lgbm= lgb.train(params= parameters, train_set= train_data, num_boost_round= 2000)
y_pred = model_lgbm.predict(x_test)
y_pred_lgbm= pd.DataFrame(y_pred)

MSE_lgbm = mean_squared_error(y_pred_lgbm,y_test)
R2_lgbm=r2_score(y_pred_lgbm,y_test)
RMSE_lgbm = np.sqrt(MSE_lgbm)


print('The RMSE for LightGBM is: ', RMSE_lgbm)
print('The MSE for LightGBM is: ', MSE_lgbm)
print('The R2 for LightGBM is: ', R2_lgbm)
```
