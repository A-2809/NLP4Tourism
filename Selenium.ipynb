```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from tqdm import tqdm
import time
import pandas as pd
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager 
```

```python
year = []
reviews = []
location = []
user_name = []
votes = []
profile = []

website = 'https://www.tripadvisor.com/SearchForums?q=new+delhi&s='

optons = Options()
# optons.add_argument("--headless")
# optons.add_experimental_option("detach", True)

driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()),
                          options = optons)

driver.get(website)

pg_number = 90

while(True):
    try:
        post_links = driver.find_elements(By.XPATH,'//a[@onclick = "setPID(34631)"]')
        for i in range(len(post_links)):
            button = driver.find_elements(By.XPATH,'//a[@onclick = "setPID(34631)"]')
            button[i].click()
            reviews_ = driver.find_elements(By.XPATH,'//div[@class ="postBody"]')
            dates = driver.find_elements(By.XPATH,'//div[@class = "postDate"]')
            loc = driver.find_elements(By.XPATH,'//div[@class = "location"]')
            name = driver.find_elements(By.XPATH,'//div[@class = "username"]')            
            profile_box = driver.find_elements(By.XPATH, '//div[@class = "profile"]')

            for j in range(len(name)):
                
                p = profile_box[j].text
               
                profile += [p]
                
                if profile == ['']:    
                    continue

                r = reviews_[j].text
                d = dates[j].text
                l = loc[j].text
                #u = name[j].text
                

                ######### for year
                try:
                    diff = int(d.split(' ')[0])
                    y = 2023 - int(diff)
                    
                except:
                    y = int(d.split(',')[1])
                

                year.append(int(y))
                reviews.append(r)
                location.append(l)
                #user_name.append(u)
                profile.append(p)

            driver.back()

        next_button = driver.find_element(By.XPATH, '//a[@href = "/SearchForums?q=new+delhi&s= &o='+ str(pg_number)+'"]')
        next_button.click()
        print(pg_number)
        pg_number = pg_number+10

    except:
        driver.quit()
        break
```
